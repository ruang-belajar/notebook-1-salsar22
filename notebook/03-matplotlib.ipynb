# Grafik Sederhana

````python
# import modul matplotlib
import matplotlib.pyplot as plt
````

## Grafik XY (_plot_)
Berikut contoh sederhana membuat visualisasi data dari _list_ menggunakan _matplotlib_

````python
y = [20, 44,21,44,65,67,29,89]
x = [2001,2002,2003,2004,2005,2006,2007,2008]

plt.plot(x,y)
plt.show()
````

### Menambahkan keterangan



````python
y = [20, 44,21,44,65,67,29,89]
x = [2001,2002,2003,2004,2005,2006,2007,2008]

plt.plot(x,y, label="Produk A")
plt.ylabel('Sales')
plt.xlabel('Tahun')
plt.title("Grafik Penjualan per Tahun")
plt.legend()
plt.show()
````

### Plotting lebih dari satu data

````python
produk_a = [20, 44,21,44,65,67,29,89]
produk_b = [30, 74,27,54,25,37,59,39]
x = [2001,2002,2003,2004,2005,2006,2007,2008]

plt.plot(x,produk_a, label="Produk A")
plt.plot(x,produk_b, label="Produk B")
plt.ylabel('Sales')
plt.xlabel('Tahun')
plt.title("Grafik Penjualan per Tahun")
plt.legend()
plt.show()
````

## Grafik _bar_

````python
produk_a = [20, 44,21,44,65,67,29,89]
x = [2001,2002,2003,2004,2005,2006,2007,2008]

plt.bar(x,produk_a, label="Produk A")
plt.show()
````

## Grafik _pie_

Untuk membuat grafik _pie_ data yang diperlukan relatif lebih sederhana, cuma dibutuhkan satu _list_

````python
data = [20,43,66,33]

plt.pie(data)
plt.show()
````

````python
data = [20,43,66,33]

judul = ['Tasik', 'Depok', 'Garut', 'Bandung']

# parameter autopct digunakan untuk menampilkan angka pada pie
# parameter labels digunakan untuk menambahkan label
plt.pie(data, autopct='%1.1f%%', labels=judul)

plt.show()
````

# Latihan

#### Jenis data seperti apa yang kita cocok tampilkan untuk tipe grafik **plot XY**?

>



1. Data yang memiliki hubungan linier. Plot XY sangat cocok untuk menampilkan  hubungan linier antara dua variabel. Hubungan linier adalah hubungan di mana perubahan nilai variabel X secara proporsional menyebabkan perubahan nilai variabel Y.
2. Data yang memiliki hubungan non-linier. Plot XY juga dapat digunakan untuk menampilkan hubungan non-linier antara dua variabel. Namun, dalam hal ini, perlu ditambahkan garis tren atau fungsi untuk memperjelas pola hubungan antara kedua variabel.
3. Data yang memiliki outliers. Plot XY dapat digunakan untuk mengidentifikasi outliers, yaitu data yang tidak sesuai dengan pola umum data lainnya. Outliers dapat diidentifikasi dengan melihat titik-titik yang berada jauh dari pola umum data lainnya.

#### Jenis data seperti apa yang kita cocok tampilkan untuk tipe grafik **bar**?

1. Data kategoris dengan nilai numerik. Grafik bar sangat cocok untuk menampilkan data kategoris dengan nilai numerik. Misalnya, grafik bar dapat digunakan untuk menampilkan data penjualan produk berdasarkan kategori produk, atau data pendapatan berdasarkan wilayah.
2. Data kategoris dengan nilai frekuensi. Grafik bar juga dapat digunakan untuk menampilkan data kategoris dengan nilai frekuensi. Misalnya, grafik bar dapat digunakan untuk menampilkan data jumlah penduduk berdasarkan agama, atau data jumlah pengunjung website berdasarkan sumber pengunjung.
3. Data kategoris dengan nilai persentase. Grafik bar juga dapat digunakan untuk menampilkan data kategoris dengan nilai persentase. Misalnya, grafik bar dapat digunakan untuk menampilkan data persentase suara yang diperoleh oleh masing-masing partai politik dalam suatu pemilihan umum.

#### Jenis data seperti apa yang kita cocok tampilkan untuk tipe grafik **pie**?

1. Pertimbangkan jumlah kategori data yang akan ditampilkan. Jika jumlah kategori datanya sedikit, maka grafik pie akan lebih efektif. Namun, jika jumlah kategori datanya banyak, maka grafik pie akan menjadi terlalu rumit dan sulit dibaca.
2. Pertimbangkan nilai data yang akan ditampilkan. Jika nilai datanya kecil, maka gunakan skala yang sesuai agar irisan-irisan grafik tidak terlalu kecil.
Pertimbangkan tujuan penggunaan grafik. Jika grafik digunakan untuk membandingkan nilai data antar kategori, maka gunakan grafik pie bertumpuk.
